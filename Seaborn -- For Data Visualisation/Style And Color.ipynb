## Style And Color

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.set_style('ticks')
sns.countplot(x='sex',data = tips)
```

```python
sns.set_style('ticks')
sns.countplot(x='sex',data = tips)
sns.despine() #by default right and top will be removed, if u want to remove left and bottom makw left = True and bottom = True as an argument
```

```python
plt.figure(figsize=(12,3))
sns.countplot(x='sex',data = tips)
```

#### Scale and Context

```python
sns.set_context('notebook',font_scale=1)
sns.countplot(x='sex',data = tips)
```

#### Palettes

```python
sns.lmplot(x='total_bill',y='tip',data = tips,hue='sex',palette='seismic')
```
