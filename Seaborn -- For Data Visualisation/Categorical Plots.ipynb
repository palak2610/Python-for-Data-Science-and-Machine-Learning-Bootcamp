## Cactegorical Plots

```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

```python
import numpy as np
```

```python
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)
```

#### Count Plot

```python
sns.countplot(x='sex',data=tips)  ##y is already being chosen here so it will just give how many male and female occurs.
```

#### Box Plot

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker')
```

#### Violin Plot

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
```

#### Strip Plot

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True,hue='sex',dodge=True) ##jitter is going to add random noise to separate the stack points
```

#### Swarm plot

```python
sns.swarmplot(x='day',y='total_bill',data=tips)

##draw back --> not very good for large datasets
```

```python
sns.violinplot(x='day',y='total_bill',data=tips)
sns.swarmplot(x='day',y='total_bill',data=tips,color='black')
```

#### Cat Plot

```python
sns.catplot(x='day',y='total_bill',data=tips,kind='bar') 
```
